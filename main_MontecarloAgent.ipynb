# Test Gridworld Environment

```python
import numpy as np
import matplotlib.pyplot as plt

from grid_world_environment import GridWorldEnvironment
from montecarloAgent import MontecarloAgent

from IPython.display import clear_output
```

## Constants

```python
MALFUNCTION_PROBABILITY = 0.0
MIN_EPSILON = 0.01
ALPHA_EPSILON = 2/3
DELAY = 0.001
DISCOUNT_FACTOR = 0.95
```

## Initialize environment

```python
plt.style.use("default")
size = (11, 11)
obstacles = [(0,3), (1,3), (2,3), (3,3), (4,3), (5,3), (6,3), (7,3),
             (3,7), (4,7), (5,7), (6,7), (7,7), (8,7), (9,7), (10,7)]
position = (0, 0)
target = (10, 10)

env = GridWorldEnvironment(size=size,
                          obstacles=obstacles,
                          position=position,
                          target=target,
                          malfunction_probability=MALFUNCTION_PROBABILITY)

env.display()
```

## 500 training episodes

```python
env.reset()
training_episodes = 500
plot_frequency = 50
mc_agent = MontecarloAgent(env, 
                           discount_factor=DISCOUNT_FACTOR,
                           min_epsilon=MIN_EPSILON,
                           alpha_epsilon=ALPHA_EPSILON,
                           max_iter=10_000)
mc_agent.train(n_episodes=training_episodes, plot=True, plot_frequency=plot_frequency)
mc_agent.plot_value_table();
```

## 10000 training episodes

```python
env.reset()
training_episodes = 100_000
plot_frequency = 100
mc_agent = MontecarloAgent(env,
                           discount_factor=DISCOUNT_FACTOR,
                           min_epsilon=MIN_EPSILON,
                           alpha_epsilon=ALPHA_EPSILON,
                           max_iter=10_000)

mc_agent.train(n_episodes=training_episodes, plot=True, plot_frequency=plot_frequency)
mc_agent.plot_value_table();
```
