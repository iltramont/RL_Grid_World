# Test Gridworld Environment

```python
import numpy as np
import matplotlib.pyplot as plt

from grid_world_environment import GridWorldEnvironment
from montecarloAgent import MontecarloAgent

from IPython.display import clear_output

np.random.seed(42)
```

## Constants

```python
MALFUNCTION_PROBABILITY = 0.0
EPSILON = 0.5
MIN_EPSILON = 0.1
ALPHA_EPSILON = 2/3
DELAY = 0.001
DISCOUNT_FACTOR = 1
Q_DISCOUNT_FACTOR = 1
```

## Initialize environment

```python
plt.style.use("default")
size = (10, 10)
obstacles = [(0,1), (0,2), (1, 1), (1,2), (2,2), (3,2), (4,2), (5,3), (6,4), (7,5), (8,6), (4,6), (3,6), (9,8), (8,8), (7,8), (6,8), (5,8), (4,8), (4,7)]
position = (0, 0)
target = (9, 9)

env = GridWorldEnvironment(size=size,
                          obstacles=obstacles,
                          position=position,
                          target=target,
                          malfunction_probability=MALFUNCTION_PROBABILITY)
env.display()
```

## 500 training episodes

```python
env.reset()
training_episodes = 500
plot_frequency = 25
mc_agent = MontecarloAgent(env, discount_factor=DISCOUNT_FACTOR, epsilon=EPSILON, min_epsilon=MIN_EPSILON, q_discount_factor=Q_DISCOUNT_FACTOR, alpha_epsilon=ALPHA_EPSILON)
mc_agent.train(n_episodes=training_episodes, plot=True, plot_frequency=plot_frequency)
mc_agent.plot_value_table();
```

## 10000 training episodes

```python
env.reset()
training_episodes = 5000
plot_frequency = 50
mc_agent = MontecarloAgent(env, discount_factor=DISCOUNT_FACTOR, epsilon=EPSILON, min_epsilon=MIN_EPSILON, q_discount_factor=Q_DISCOUNT_FACTOR, alpha_epsilon=ALPHA_EPSILON)
mc_agent.train(n_episodes=training_episodes, plot=True, plot_frequency=plot_frequency)
mc_agent.plot_value_table();
```
